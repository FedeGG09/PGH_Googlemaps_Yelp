```python
import seaborn as sns
```

```python
# Filtrar la columna 'name' por los nombres especificados
filtro_nombres = ['McDonald\'s', 'Starbucks', 'Subway', 'KFC', 'Domino\'s Pizza']
df_filtrado = df[df['name'].isin(filtro_nombres)]

# Ordenar el DataFrame por cantidad de estrellas (stars) de mayor a menor
df_filtrado = df_filtrado.sort_values(by='stars_y', ascending=False)

# Crear un gráfico de barras apiladas para mostrar las calificaciones por restaurante
plt.figure(figsize=(10, 6))
sns.barplot(x='stars_y', y='name', data=df_filtrado, ci=None, palette='Set3')
plt.xlabel('Calificación (Stars)')
plt.ylabel('Restaurante')
plt.title('Calificaciones de Restaurantes (Ordenados por Estrellas)')
plt.show()
```

```python
# Filtrar la columna 'name' por los nombres especificados
filtro_nombres = ['McDonald\'s', 'Starbucks', 'Subway', 'KFC', 'Domino\'s Pizza']
df_filtrado = df[df['name'].isin(filtro_nombres)]

# Calcular el promedio de calificaciones para cada restaurante
promedios_por_restaurante = df_filtrado.groupby('name')['stars_y'].mean().reset_index()

# Ordenar el DataFrame por promedio de calificación de mayor a menor
promedios_por_restaurante = promedios_por_restaurante.sort_values(by='stars_y', ascending=False)

# Crear un gráfico de barras horizontal para mostrar los promedios de calificaciones ordenados
plt.figure(figsize=(10, 6))
sns.barplot(x='stars_y', y='name', data=promedios_por_restaurante, ci=None, palette='Set3')
plt.xlabel('Promedio de Calificación')
plt.ylabel('Restaurante')
plt.title('Promedio de Calificaciones por Restaurante (Ordenado de Mayor a Menor)')

# Mostrar una tabla con los promedios de calificaciones por restaurante
plt.figure(figsize=(8, 4))
table = plt.table(cellText=promedios_por_restaurante.values, colLabels=['Restaurante', 'Promedio de Calificación'], cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 1.5)
plt.axis('off')
plt.show()
```

```python
# Filtrar la columna 'name' por los nombres especificados
filtro_nombres = ['McDonald\'s', 'Starbucks', 'Subway', 'KFC', 'Domino\'s Pizza']
df_filtrado = df[df['name'].isin(filtro_nombres)]

# Definir una paleta de colores diferente para cada cadena de restaurantes
paleta_colores = sns.color_palette("husl", len(filtro_nombres))

# Crear un gráfico de barras para cada cadena por estado con colores separados
plt.figure(figsize=(12, 8))
for i, nombre in enumerate(filtro_nombres):
    df_nombre = df_filtrado[df_filtrado['name'] == nombre]
    sns.barplot(x='state', y='stars_y', data=df_nombre, ci=None, label=nombre, color=paleta_colores[i])

plt.xlabel('Estado')
plt.ylabel('Promedio de Calificación')
plt.title('Promedio de Calificaciones por Cadena de Restaurantes y Estado')
plt.legend(title='Cadena de Restaurantes')
plt.show()
```

```python
# Nombres de cadenas de restaurantes a considerar
nombres_cadenas = ['McDonald\'s', 'Starbucks', 'Subway', 'KFC', 'Domino\'s Pizza']

# Definir una paleta de colores personalizada
colores = ['blue', 'green', 'red', 'purple', 'orange']

# Crear un histograma separado para cada cadena de restaurantes con colores personalizados
plt.figure(figsize=(12, 8))

for i, nombre_cadena in enumerate(nombres_cadenas):
    plt.hist(df[df['name'] == nombre_cadena]['stars_y'], bins=5, edgecolor='k', alpha=0.5, label=nombre_cadena, color=colores[i])

plt.xlabel('Calificación (Stars)')
plt.ylabel('Número de Restaurantes')
plt.title('Distribución de Calificaciones de Restaurantes por Cadena')
plt.legend(title='Cadena de Restaurantes')
plt.show()
```

```python
# Nombres de cadenas de restaurantes a considerar
nombres_cadenas = ['McDonald\'s', 'Starbucks', 'Subway', 'KFC', 'Domino\'s Pizza']

# Filtrar el DataFrame por las cadenas de restaurantes
df_filtrado = df[df['name'].isin(nombres_cadenas)]

# Crear un gráfico de caja separado para cada cadena de restaurantes con paleta de colores 'Set3'
plt.figure(figsize=(12, 6))
sns.boxplot(x='state', y='stars_y', data=df_filtrado, palette='Set3', hue='name')
plt.xlabel('Estado')
plt.ylabel('Calificación (Stars)')
plt.title('Distribución de Calificaciones por Estado y Cadena de Restaurantes')
plt.xticks(rotation=45)
plt.legend(title='Cadena de Restaurantes')
plt.show()
```
