```python
# Filtrar la columna 'name' por los nombres especificados
filtro_nombres = ['McDonald\'s', 'Starbucks', 'Subway', 'KFC', 'Domino\'s Pizza']
df_filtrado = df[df['name'].isin(filtro_nombres)]

# Calcular el promedio de calificaciones para cada restaurante
promedios_por_restaurante = df_filtrado.groupby('name')['stars_y'].mean().reset_index()

# Ordenar el DataFrame por promedio de calificación de mayor a menor
promedios_por_restaurante = promedios_por_restaurante.sort_values(by='stars_y', ascending=False)

# Crear un gráfico de barras horizontal para mostrar los promedios de calificaciones ordenados
plt.figure(figsize=(10, 6))
sns.barplot(x='stars_y', y='name', data=promedios_por_restaurante, ci=None, palette='Set3')
plt.xlabel('Promedio de Calificación')
plt.ylabel('Restaurante')
plt.title('Promedio de Calificaciones por Restaurante (Ordenado de Mayor a Menor)')

# Mostrar una tabla con los promedios de calificaciones por restaurante
plt.figure(figsize=(8, 4))
table = plt.table(cellText=promedios_por_restaurante.values, colLabels=['Restaurante', 'Promedio de Calificación'], cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 1.5)
plt.axis('off')
plt.show()
```

```python
# Filtrar la columna 'name' por los nombres especificados
filtro_nombres = ['McDonald\'s', 'Starbucks', 'Subway', 'KFC', 'Domino\'s Pizza']
df_filtrado = df[df['name'].isin(filtro_nombres)]

# Definir una paleta de colores diferente para cada cadena de restaurantes
paleta_colores = sns.color_palette("husl", len(filtro_nombres))

# Crear un gráfico de barras para cada cadena por estado con colores separados
plt.figure(figsize=(12, 8))
for i, nombre in enumerate(filtro_nombres):
    df_nombre = df_filtrado[df_filtrado['name'] == nombre]
    sns.barplot(x='state', y='stars_y', data=df_nombre, ci=None, label=nombre, color=paleta_colores[i])

plt.xlabel('Estado')
plt.ylabel('Promedio de Calificación')
plt.title('Promedio de Calificaciones por Cadena de Restaurantes y Estado')
plt.legend(title='Cadena de Restaurantes')
plt.show()
```

```python
# Calcular el total de review_count por cadena
total_review_count_por_cadena = df_filtrado.groupby('name')['review_count'].sum()

# Crear un nuevo DataFrame con los resultados
df_total_review_count = pd.DataFrame({
    'Cadena': total_review_count_por_cadena.index,
    'Total_Review_Count': total_review_count_por_cadena.values
})

# Ordenar el DataFrame por 'Total_Review_Count' de mayor a menor
df_total_review_count = df_total_review_count.sort_values(by='Total_Review_Count', ascending=False)

# Mostrar el DataFrame ordenado con display
display(df_total_review_count)
```

```python
import pandas as pd

# Agrupar y contar la cantidad de restaurantes por cadena
restaurantes_por_cadena = df_filtrado['name'].value_counts().reset_index()
restaurantes_por_cadena.columns = ['Cadena', 'Cantidad de Restaurantes']

# Mostrar la tabla
print(restaurantes_por_cadena)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame df_filtrado

# Contar la cantidad de restaurantes por cadena
restaurantes_por_cadena = df_filtrado['name'].value_counts().reset_index()
restaurantes_por_cadena.columns = ['Cadena', 'Cantidad de Restaurantes']

# Calcular el porcentaje para cada cadena
total_restaurantes = restaurantes_por_cadena['Cantidad de Restaurantes'].sum()
restaurantes_por_cadena['Porcentaje'] = (restaurantes_por_cadena['Cantidad de Restaurantes'] / total_restaurantes) * 100

# Mostrar la tabla con porcentajes
print(restaurantes_por_cadena)

# Crear un gráfico de torta con etiquetas y porcentajes
plt.figure(figsize=(8, 8))
plt.pie(restaurantes_por_cadena['Porcentaje'], labels=restaurantes_por_cadena['Cadena'], autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Restaurantes por Cadena')
plt.axis('equal')  # Hace que el gráfico de torta sea circular
plt.show()
```

```python
#Calcular la cantidad de restaurantes por cadena y estado
restaurantes_por_cadena_estado = df_filtrado.groupby(['name', 'state']).size().reset_index(name='Cantidad de Restaurantes')

# Crear una tabla pivote para mostrar la cantidad por estado
pivot_table = pd.pivot_table(restaurantes_por_cadena_estado, values='Cantidad de Restaurantes', index='state', columns='name', aggfunc='sum', fill_value=0)

# Mostrar la tabla pivote
display(pivot_table)
```

```python
# Calcular la calificación promedio y el número total de restaurantes por cadena
calificacion_promedio_por_cadena = df_filtrado.groupby('name')['stars_y'].mean().reset_index()
total_restaurantes_por_cadena = df_filtrado['name'].value_counts().reset_index()
calificacion_promedio_por_cadena.columns = ['Cadena', 'Calificación Promedio']
total_restaurantes_por_cadena.columns = ['Cadena', 'Total Restaurantes']

# Combinar los DataFrames de calificación promedio y total de restaurantes
datos_por_cadena = pd.merge(calificacion_promedio_por_cadena, total_restaurantes_por_cadena, on='Cadena')

# Ordenar las cadenas por calificación promedio
datos_por_cadena = datos_por_cadena.sort_values(by='Calificación Promedio', ascending=False)

# Crear un gráfico de barras con colores diferentes
fig, ax1 = plt.subplots(figsize=(12, 6))
color = 'tab:blue'
ax1.bar(datos_por_cadena['Cadena'], datos_por_cadena['Calificación Promedio'], color=color, alpha=0.7)
ax1.set_xlabel('Cadena')
ax1.set_ylabel('Calificación Promedio (stars_y)', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Crear un segundo eje y para el número total de restaurantes
ax2 = ax1.twinx()
color = 'tab:orange'
ax2.bar(datos_por_cadena['Cadena'], datos_por_cadena['Total Restaurantes'], color=color, alpha=0.7)
ax2.set_ylabel('Total de Restaurantes', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Añadir números en las barras
for i, row in enumerate(datos_por_cadena.iterrows()):
    cadena, datos = row
    plt.text(i, datos['Calificación Promedio'] + 0.1, f'{int(datos["Total Restaurantes"])}', ha='center', fontsize=10)

# Rotar las etiquetas de las cadenas para mejorar la legibilidad
plt.xticks(rotation=90)
plt.title('Calificación Promedio y Total de Restaurantes por Cadena')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
