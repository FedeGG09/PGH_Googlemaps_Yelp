# Se parte el archivo review.json en partes para poder trabajarlo en local

```python
import pandas as pd
import os

# Ruta al archivo JSON grande
input_json_file = "C:/Users/Fede/Documents/Henry/Proyecto 3/Yelp/review.json"

# Directorio donde se guardarán las partes divididas
output_directory = "C:/Users/Fede/Documents/Henry/Proyecto 3/Yelp/review_parts"

# Crea el directorio de salida si no existe
os.makedirs(output_directory, exist_ok=True)

# Tamaño máximo de cada parte (ajusta según tus necesidades)
max_chunk_size = 200000  # por ejemplo, dividir en partes de 500,000 filas

# Lee el archivo JSON grande en un DataFrame de pandas
df = pd.read_json(input_json_file, lines=True, chunksize=max_chunk_size)

part_number = 0  # Inicializa el número de parte

for chunk in df:
    part_number += 1  # Incrementa el número de parte en cada iteración
    output_file = os.path.join(output_directory, f"part_{part_number}.json")
    chunk.to_json(output_file, orient="records", lines=True)

```

```python
import pandas as pd
import datetime as dt
import demoji
import re
from datetime import datetime
from IPython.display import display

# Cargar
df_yelp = pd.read_csv('C:/Users/Fede/Documents/Henry/Proyecto 3/etl/YelpReviewyBusiness2.csv')

# Seleccionar columnas
selected_columns = ['review_id', 'user_id', 'business_id', 'stars_y', 'text', 'date']
df_yelp = df_yelp[selected_columns]

# Elimina nulos en la columna 'text'
df_yelp.dropna(subset=['text'], inplace=True)

# Cambiar a formato fecha, creando columnas para año, mes y día
df_yelp['date'] = pd.to_datetime(df_yelp['date'])
df_yelp['anio'] = df_yelp['date'].dt.year
df_yelp['mes'] = df_yelp['date'].dt.month
df_yelp['dia'] = df_yelp['date'].dt.day
df_yelp.drop(['date'], axis=1, inplace=True)

# Selección de registros desde 2017 hasta 2021
df_yelp = df_yelp[df_yelp['anio'] >= 2017]

# Limpieza de datos en la columna 'text': a minúsculas, eliminar emojis y signos de puntuación
df_yelp['text'] = df_yelp['text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df_yelp['text'] = df_yelp['text'].apply(lambda x: demoji.replace(x, ''))
df_yelp['text'] = df_yelp['text'].str.lower()

# Guardar el DataFrame limpio en un archivo CSV sin index
df_yelp.to_csv('YelpReviewBusinessLimpio.csv', index=False)

# Mostrar el DataFrame
display(df_yelp)
```
