# Buscamos que estÃ©n las cadenas de restaurants con las que vamos a trabajar.

```python
import pandas as pd
from fuzzywuzzy import fuzz, process

# Ruta al archivo CSV
ruta_csv = "C:/Users/Fede/Desktop/ETL/3er_Demo/YelpEstados.csv"

# Cargar el archivo CSV en un DataFrame
restaurantes_df = pd.read_csv(ruta_csv)

# Cambiar el tipo de dato de la columna 'Negocio' a str
restaurantes_df['Negocio'] = restaurantes_df['Negocio'].astype(str)

# Crear una lista con las cadenas de restaurantes que deseas buscar
cadenas_de_restaurantes = ["KFC", "McDonald's", "Subway", "Starbucks", "Domino's Pizza"]

# Crear un diccionario para almacenar los recuentos de cada cadena
recuentos = {}

# Iterar sobre la lista de cadenas de restaurantes y buscar coincidencias difusas en el DataFrame
for cadena in cadenas_de_restaurantes:
    # Utilizar process.extract para encontrar las coincidencias mÃ¡s cercanas en el DataFrame
    coincidencias = process.extract(cadena, restaurantes_df['Negocio'], limit=10, scorer=fuzz.ratio)
    
    # Filtrar las coincidencias que superan un umbral de similitud (por ejemplo, 80)
    coincidencias_filtradas = [match for match in coincidencias if match[1] >= 80]
    
    # Contar cuÃ¡ntas veces aparece la cadena o las coincidencias similares en el DataFrame
    recuento = len(coincidencias_filtradas)
    recuentos[cadena] = recuento

# Imprimir los recuentos
for cadena, recuento in recuentos.items():
    print(f"{cadena}: {recuento} veces")
```
