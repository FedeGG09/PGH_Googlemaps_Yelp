# Se obtiene el rango de precios sacado de la columna attributes, siendo 1 Accesible, 2 Moderado, 3 Costoso y 4 Muy Caro y se lo agrega a una nueva columna.

```python
import pandas as pd
from textblob import TextBlob

# Ruta al archivo CSV
archivo_csv = "C:/Users/Fede/Desktop/ETL/3er_Demo/YelpEstados_Nombres.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(archivo_csv)

# Limpia los valores nulos en la columna "Review" y convierte todo a cadenas
df['Review'] = df['Review'].fillna('').astype(str)

# Crear una función para aplicar el análisis de sentimiento con TextBlob
def get_sentiment_score(review):
    analysis = TextBlob(review)
    return analysis.sentiment.polarity  # Usamos la polaridad para obtener un valor entre -1 y 1

# Aplicar la función a la columna "Review" y almacenar los resultados en una nueva columna "Sentiment_Score"
df['Sentiment_Score'] = df['Review'].apply(get_sentiment_score)

# Ver el DataFrame con la columna de puntuaciones de sentimiento
df
```

# Se obtiene el rango de precios sacado de la columna attributes, siendo 1 Accesible, 2 Moderado, 3 Costoso y 4 Muy Caro y se lo agrega a una nueva columna.

```python
import pandas as pd

# Ruta al archivo CSV
archivo_csv = "C:/Users/Fede/Desktop/ETL/3er_Demo/YELP_Recomendaciones.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(archivo_csv)

# Función para obtener el rango de precios numérico
def obtener_rango_precios(attributes):
    try:
        # Convierte la cadena de atributos en un diccionario
        attributes_dict = eval(attributes)
        
        # Busca las claves relacionadas con los rangos de precios y obtén su valor
        for key in ['RestaurantsPriceRange1', 'RestaurantsPriceRange2', 'RestaurantsPriceRange3', 'RestaurantsPriceRange4']:
            if key in attributes_dict:
                return int(attributes_dict[key])
        
        # Si no se encuentra ninguna clave de rango de precios, devuelve None
        return None
    except:
        return None

# Aplica la función a la columna "attributes" y almacena los resultados en una nueva columna "Rango_Precios"
df['Rango_Precios'] = df['attributes'].apply(obtener_rango_precios)

# Muestra el DataFrame con la nueva columna "Rango_Precios"
print(df[['Negocio', 'attributes', 'Rango_Precios']])
```

# Realizamos una búsqueda de los mejores 20 locales para obtener los mejores calificados y en que se diferencian según sus atributos. 

```python
# Ordenar el DataFrame según Estrellas_Negocio en orden descendente y tomar los 20 mejores
top_20_estrellas = df.sort_values(by="Estrellas_Negocio", ascending=False).head(20)

# Ordenar el DataFrame según Rating_Usuario en orden descendente y tomar los 20 mejores
top_20_rating_usuario = df.sort_values(by="Rating_Usuario", ascending=False).head(20)

# Ordenar el DataFrame según Numero_reviews en orden descendente y tomar los 20 mejores
top_20_numero_reviews = df.sort_values(by="Numero_reviews", ascending=False).head(20)

# Ordenar el DataFrame según Sentiment_Score en orden descendente y tomar los 20 mejores
top_20_sentiment_score = df.sort_values(by="Sentiment_Score", ascending=False).head(20)
```

```python
# Concatenar los cuatro DataFrames en uno solo
top_20_combined = pd.concat([top_20_estrellas, top_20_rating_usuario, top_20_numero_reviews, top_20_sentiment_score])

# Eliminar duplicados (si hay alguno)
top_20_combined = top_20_combined.drop_duplicates()

# Mostrar el DataFrame resultante
top_20_combined
```

# Se exploran los atributos destacados en los restaurantes mejor calificados segun su número de estrellas y la calificación de los usuarios.

```python
import pandas as pd
import re
import ast

# Ruta al archivo CSV
archivo_csv = "C:/Users/Fede/Desktop/ETL/3er_Demo/top_20_combined.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(archivo_csv)

# Función para convertir cadenas de texto en diccionarios
def parse_dict(text):
    try:
        # Encuentra todas las claves y valores dentro de la cadena
        matches = re.findall(r"'(.*?)': (?:'([^']*)'|(True|False|u'[^']*'))", text)
        
        # Construye un diccionario a partir de las claves y valores encontrados
        result = {k: v if v != "" else (True if bool_val == 'True' else False) for k, v, bool_val in matches}
        return result
    except:
        return {}

# Aplicar la función a la columna "attributes"
df['attributes'] = df['attributes'].apply(parse_dict)

# Crear un conjunto (set) para almacenar todas las características únicas
unique_features = set()

# Iterar a través de los diccionarios y agregar las características al conjunto
for attributes_dict in df['attributes']:
    unique_features.update(attributes_dict.keys())

# Convertir el conjunto de características únicas en una lista
unique_features_list = list(unique_features)

# Mostrar las características comunes
print(unique_features_list)
```

# Se crea una nueva columna llamada Atributos Deseados con el porcentaje de la coincidencia de los restaurantes con los atributos seleccionados.

```python
import pandas as pd

# Cargar el archivo CSV
archivo_csv = "C:/Users/Fede/Desktop/ETL/3er_Demo/YELP_Recomendaciones.csv"
df = pd.read_csv(archivo_csv)

# Definir las características para buscar
caracteristicas = [
    'upscale: None', 'OutdoorSeating', 'valet', 'NoiseLevel: u\'average\'',
    'NoiseLevel: u\'quiet\'', 'HappyHour', 'RestaurantsTakeOut', 'WiFi: u\'free\'',
    'GoodForKids', 'Caters', 'Smoking: u\'no\'', 'RestaurantsGoodForGroups',
    'trendy: None', 'intimate', 'HasTV', 'BusinessAcceptsCreditCards', 'dessert',
    'WheelchairAccessible', 'validated: la validación de estacionamiento en un establecimiento',
    'lot', 'video', 'BikeParking', 'background_music', 'karaoke', 'jukebox', 'DogsAllowed'
]

# Función para calcular el porcentaje de coincidencia
def calcular_porcentaje_coincidencia(attributes):
    if isinstance(attributes, str):
        coincidencias = 0
        for caracteristica in caracteristicas:
            if caracteristica in attributes:
                coincidencias += 1
        return (coincidencias / len(caracteristicas)) * 100
    else:
        return 0  # Manejar casos en los que no hay atributos

# Aplicar la función y crear una nueva columna "PorcentajeCoincidencia"
df['Atributos Deseados'] = df['attributes'].apply(calcular_porcentaje_coincidencia)

df
```
