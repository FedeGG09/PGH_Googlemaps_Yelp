```python
import pandas as pd

# Ruta al archivo CSV
archivo_csv = "C:/Users/Fede/Desktop/ETL/3er_Demo/YELP_Recomendaciones.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(archivo_csv)

# Mostrar el DataFrame con el widget "display"
display(df)
```

# Buscamos el sentimiento promedio.

```python
sentimiento_promedio = df['Sentiment_Score'].mean()
print(f"Sentimiento promedio: {sentimiento_promedio}")
```

# Filtramos el DataFrame por las opiniones negativas.

```python
opiniones_negativas = df[df['Sentiment_Score'] < 0]
opiniones_negativas
```

# Buscamos las palabras clave mas comunes entre las opiniones negativas.

```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk import word_tokenize, pos_tag

# Combinar todas las opiniones negativas en un solo texto
opiniones_negativas_texto = ' '.join(opiniones_negativas['Review'])

# Tokenización y eliminación de stopwords
palabras_clave = [word.lower() for word in word_tokenize(opiniones_negativas_texto) if word.lower() not in stopwords.words('english')]

# Etiquetar las palabras con sus partes del discurso
palabras_con_pos = pos_tag(palabras_clave)

# Encontrar adjetivos calificativos (JJ) y frases sustantivo-adjetivo (NN JJ)
adjetivos_y_frases = [word for word, pos in palabras_con_pos if pos in ['JJ', 'NNS', 'NN', 'JJR', 'JJS']]

# Contar las palabras clave más comunes
palabras_clave_contadas = Counter(adjetivos_y_frases)

# Imprimir las palabras clave más comunes
print(palabras_clave_contadas.most_common(50))
```

# Identificamos las entidades (como nombres de lugares, productos, etc.) en las opiniones negativas con SpaCy para identificar problemas específicos en las opiniones negativas y comprender mejor los temas recurrentes en las opiniones de los clientes.

```python
# Importar spaCy
import spacy

# Cargar el modelo en inglés
nlp = spacy.load("en_core_web_sm")

# Analizar cada opinión negativa y extraer entidades
entidades_negativas = []

for opinion in opiniones_negativas['Review']:
    doc = nlp(opinion)
    for entidad in doc.ents:
        entidades_negativas.append(entidad.text)

# Contar las entidades más comunes
entidades_frecuentes = Counter(entidades_negativas)

# Imprimir las entidades más comunes
print(entidades_frecuentes.most_common(30))
```
